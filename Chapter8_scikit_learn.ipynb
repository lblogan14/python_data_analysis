#The Iris Flower Dataset
Data from three different species of iris (Iris silky, virginica Iris, and Iris versicolor),

correspond to the *length* and *width* of the *sepals* and the *length* and *width* of the *petals*

```
from sklearn import datasets
iris = datasets.load_iris()
```

```
iris.data
```

```
iris.target
```

```
iris.target_names
```

To better understand the dataset, use the matplotlib library, using a scatterplot,

xaxis - length of the sepal,

yaxis - width of the sepal.

```
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```
x = iris.data[:,0] #x-axis - sepal length
y = iris.data[:,1] #y-aixs - sepal width
species = iris.target #Species
#convert numbers to color kwargs
color = {0:'r', 1:'b', 2:'g'}
c = [color[i] for i in species]

#prepare for the figure area
x_min, x_max = x.min()-0.5, x.max()+0.5
y_min,y_max = y.min()-0.5, y.max()+0.5

#Scatterplot
plt.figure
plt.title('Iris Dataset - Classification By Sepal Sizes')
plt.scatter(x,y, c=c) # c = color,sequence, or seq of color
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

Use the same procedure for the length and the width of the petals

```
x = iris.data[:,2] #X-Axis - petal length
y = iris.data[:,3] #Y-Axis - petal length
species = iris.target

x_min, x_max = x.min() - .5,x.max() + .5
y_min, y_max = y.min() - .5,y.max() + .5

#SCATTERPLOT
plt.figure()
plt.title('Iris Dataset - Classification By Petal Sizes', size=14)
plt.scatter(x,y, c=c)
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

#The PCA Decomposition
To unify the two scatterplots in one scatterplot, you need to reduce all these features into 3 principal components for the 3D scatterplot.

```
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
x_reduced = pca.fit_transform(iris.data)
```

To use the 3D scatterplot, you need to import **Axes3D** from **mpl_toolkits.mplot3d** module

```
from mpl_toolkits.mplot3d import Axes3D
```

```
#Scatterplot 3D
fig = plt.figure()
ax = Axes3D(fig)
ax.set_title('Iris Dataset by PCA', size=15)
ax.scatter(x_reduced[:,0], x_reduced[:,1], x_reduced[:,2], c=c)
ax.set_xlabel('First Eigenvector')
ax.set_ylabel('Second Eigenvector')
ax.set_zlabel('Third Eigenvector')

ax.w_xaxis.set_ticklabels(())
ax.w_yaxis.set_ticklabels(())
ax.w_zaxis.set_ticklabels(())
```

Visualize the decision boundaries in a 2D scatterplot of sepals:

```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
iris = datasets.load_iris()
x = iris.data[:,:2] #X-Axis - sepal length-width
y = iris.target #Y-Axis - species
x_min, x_max = x[:,0].min() - .5,x[:,0].max() + .5
y_min, y_max = x[:,1].min() - .5,x[:,1].max() + .5

color = {0:'r', 1:'b', 2:'g'}
c = [color[i] for i in y]
```

```
#MESH
cmap_light = ListedColormap(['#AAAAFF','#AAFFAA','#FFAAAA'])
h = .02
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
#               ^create a grid
```

```
knn = KNeighborsClassifier()
knn.fit(x,y)
desicion_map = knn.predict(np.c_[xx.ravel(), yy.ravel()]) #predict all mesh points
#                         ^concatenate two matrices
#    ravel() - return a 1D array
desicion_map = desicion_map.reshape(xx.shape)
```

```
#plot decision boundaries
plt.figure()
plt.pcolormesh(xx,yy, desicion_map, cmap=cmap_light)
#                      ^ C=desicion_map = scalar 2D array, which will be color-mapped

#plot training points
plt.scatter(x[:,0], x[:,1], c=c)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
```

Same thing for the size of the petals

```
x = iris.data[:,2:4] # length and width of petals
x_min, x_max = x[:,0].min()-0.5, x[:,0].max()+0.5
y_min, y_max = x[:,1].min()-0.5, x[:,1].max()+0.5

#MESH
h = 0.02
xx, yy = np.meshgrid(np.arange(x_min,x_max,h), np.arange(y_min, y_max,h))
knn2 = KNeighborsClassifier()
knn2.fit(x,y)
decision_map = knn2.predict(np.c_[xx.ravel(), yy.ravel()])
decision_map = decision_map.reshape(xx.shape)

plt.figure()
plt.pcolormesh(xx,yy, decision_map, cmap=cmap_light)
plt.scatter(x[:,0], x[:,1], c=c)
plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
```

Now you can use all features to create 10 models

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import datasets
diabetes = datasets.load_diabetes()

x_train = diabetes.data[:-20]
y_train = diabetes.target[:-20]
x_test = diabetes.data[-20:]
y_test = diabetes.target[-20:]
```

```
plt.figure(figsize=(8,12))

for f in range(0,10):
  xi_test = x_test[:,f]
  xi_train = x_train[:,f]
  xi_test = xi_test[:,np.newaxis]
  xi_train = xi_train[:,np.newaxis]
  lr = linear_model.LinearRegression()
  lr.fit(xi_train, y_train)
  y_pred = lr.predict(xi_test)
  
  plt.subplot(5,2,f+1)
  plt.scatter(xi_test,y_test,color='k')
  plt.plot(xi_test, y_pred, color='b', linewidth=5)
```

##Support Vector Classification (SVC)

```
x = np.array([[1,3],[1,2],[1,1.5],[1.5,2],[2,3],[2.5,1.5],
[2,1],[3,1],[3,2],[3.5,1],[3.5,3]])
y = ['r']*6 + ['b']*5
plt.scatter(x[:,0],x[:,1],c=y,s=50,alpha=0.9)
```

Use the **SVC()** for classification

```
from sklearn import svm
svc = svm.SVC(kernel='linear').fit(x,y)
X,Y = np.mgrid[0:4:200j, 0:4:200j]
X
```

```
Y
```

```
#create decision boundary
Z = svc.decision_function(np.c_[X.ravel(), Y.ravel()])
Z = Z.reshape(X.shape)
```

```
#plot filled colormap
plt.contourf(X,Y, Z>0, alpha=0.4)
#plot decision boundary
plt.contour(X,Y,Z, color='k', linestyles='-', levels=0)
plt.scatter(x[:,0], x[:,1], c=y, s=50, alpha=0.9)
```

Identify the **support_vectors**

```
x = np.array([[1,3],[1,2],[1,1.5],[1.5,2],[2,3],[2.5,1.5],
[2,1],[3,1],[3,2],[3.5,1],[3.5,3]])
y = [0]*6 + [1]*5
c_index = {0:'y',1:'b'}
c = [c_index[i] for i in y]

svc = svm.SVC(kernel='linear',C=1).fit(x,y)
X,Y = np.mgrid[0:4:200j,0:4:200j]
Z = svc.decision_function(np.c_[X.ravel(),Y.ravel()])
Z = Z.reshape(X.shape)

plt.contourf(X,Y, Z>0, alpha=0.2)
plt.contourf(X,Y, Z<0, alpha=0.8)
#plot desicion boundary and separation lines
plt.contour(X,Y,Z,colors=['k','k','k'], linestyles=['--','-','--'],
            levels=[-1,0,1])
#plot support vectors
plt.scatter(svc.support_vectors_[:,0],svc.support_vectors_[:,1],s=120,
            facecolors='r')
plt.scatter(x[:,0],x[:,1],c=c,s=50,alpha=0.9)
```

If you change the regulation factor C to a small number, eg. **C=0.1**, you can see how it affect the decision boundary below:

```
svc = svm.SVC(kernel='linear',C=0.1).fit(x,y)
X,Y = np.mgrid[0:4:200j,0:4:200j]
Z = svc.decision_function(np.c_[X.ravel(),Y.ravel()])
Z = Z.reshape(X.shape)

plt.contourf(X,Y, Z>0, alpha=0.2)
plt.contourf(X,Y, Z<0, alpha=0.8)
#plot desicion boundary and separation lines
plt.contour(X,Y,Z,colors=['k','k','k'], linestyles=['--','-','--'],
            levels=[-1,0,1])
#plot support vectors
plt.scatter(svc.support_vectors_[:,0],svc.support_vectors_[:,1],s=120,
            facecolors='r')
plt.scatter(x[:,0],x[:,1],c=c,s=50,alpha=0.9)
```

##Plotting Different SVM Classifiers Using the Iris Dataset

```
iris = datasets.load_iris()
x = iris.data[:,:2]
y = iris.target
c_index = {0:'purple',1:'yellow', 2:'blue'}
c = [c_index[i] for i in y]

x_min,x_max = x[:,0].min() - .5, x[:,0].max() + .5
y_min,y_max = x[:,1].min() - .5, x[:,1].max() + .5
h = .02
X, Y = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min,y_max,h))
```

For a linear kernal,

```
svc = svm.SVC(kernel='linear',C=1.0).fit(x,y)
Z = svc.predict(np.c_[X.ravel(),Y.ravel()])
Z = Z.reshape(X.shape)
plt.contourf(X,Y,Z,alpha=0.4)
plt.contour(X,Y,Z,colors='r')
plt.scatter(x[:,0],x[:,1],c=c)
```

For a polynomial kernel,

```
svc = svm.SVC(kernel='poly',C=1.0).fit(x,y)
Z = svc.predict(np.c_[X.ravel(),Y.ravel()])
Z = Z.reshape(X.shape)
plt.contourf(X,Y,Z,alpha=0.4)
plt.contour(X,Y,Z,colors='r')
plt.scatter(x[:,0],x[:,1],c=c)
```

For a rbf kernel,

```
svc = svm.SVC(kernel='rbf',C=1.0, gamma=3).fit(x,y)
Z = svc.predict(np.c_[X.ravel(),Y.ravel()])
Z = Z.reshape(X.shape)
plt.contourf(X,Y,Z,alpha=0.4)
plt.contour(X,Y,Z,colors='r')
plt.scatter(x[:,0],x[:,1],c=c)
```

##Support Vector Regression (SVR)

```
diabetes = datasets.load_diabetes()
x_train = diabetes.data[:-20]
y_train = diabetes.target[:-20]
x_test = diabetes.data[-20:]
y_test = diabetes.target[-20:]
x0_test = x_test[:,2]
x0_train = x_train[:,2]
x0_test = x0_test[:,np.newaxis]
x0_train = x0_train[:,np.newaxis]
```

```
x0_test.sort(axis=0)
x0_test = x0_test*100
x0_train = x0_train*100
```

```
svr = svm.SVR(kernel='linear',C=1000)
svr2 = svm.SVR(kernel='poly',C=1000,degree=2)
svr3 = svm.SVR(kernel='poly',C=1000,degree=3)
svr.fit(x0_train,y_train)
svr2.fit(x0_train,y_train)
svr3.fit(x0_train,y_train)
y = svr.predict(x0_test)
y2 = svr2.predict(x0_test)
y3 = svr3.predict(x0_test)
plt.scatter(x0_test,y_test,color='k')
plt.plot(x0_test,y,color='b')
plt.plot(x0_test,y2,c='r')
plt.plot(x0_test,y3,c='g')
```
